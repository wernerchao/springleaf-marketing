```python
num_feat = train.dtypes[train.dtypes == 'float64'].index.values.tolist()
cat_feat = train.dtypes[train.dtypes == 'object'].index.values.tolist()
int_feat = train.dtypes[train.dtypes == 'int64'].index.values.tolist()
print len(num_feat), len(cat_feat)
print len(int_feat)
```

```python
# Split the data into categorical & numerical features
numerical = []
categorical = []
for col_name in int_feat:
    if len(train[col_name].unique()) > 20: # If > 20 then it's probably a numerical feature
        num_feat.append(col_name)
    else:                                  # Else then probably categorical feature
        cat_feat.append(col_name)

print len(num_feat) # 995 numerical features in int64 + 470 float features
print len(cat_feat)  # 409 categorical features in int64 + 35 object features
```

```python
# Check no. of missing values in each numerical feature.
no_cols = 0
for i, column in enumerate(num_feat):
    missing = train[column].isnull().values.sum()
    print missing
    if missing > 0:
        no_cols += 1
print 'Total no. of cols: ', i
print 'No. of cols with missing values: ', no_cols, 'columns.'
```

```python
# A function that filles in missing values with most frequent value, median, or custom value.
def feature_process(process_func='None', most_frequent_func=False, median_func=False, col_name=[]):
    print 'processing features...'
    for column in col_name:
        if median_func:
            process_func = train[column].median()
        if most_frequent_func:
            process_func = train[column].value_counts().index.values[0]
        train.loc[train[column].isnull(), column] = process_func
```

```python
# Fill numerical features missing values with median()
feature_process(median_func=True, col_name=num_feat)

# Check the result.
no_cols = 0
for i, column in enumerate(num_feat):
    missing = train[column].isnull().values.sum()
    if missing > 0:
        no_cols += 1
print 'Total no. of cols: ', i
print 'No. of cols with missing values: ', no_cols, 'columns.'
```

```python
# Check no. of missing values in each categorical feature.
no_cols = 0
for i, column in enumerate(cat_feat):
    missing = train[column].isnull().values.sum()
    print missing
    if missing > 0:
        no_cols += 1
print 'Total no. of cols: ', i
print 'No. of cols with missing values: ', no_cols, 'columns.'
```

```python
# Fill in the missing values with the most common value.
feature_process(most_frequent_func=True, col_name=cat_feat)

# Check the result.
no_cols = 0
for i, column in enumerate(cat_feat):
    missing = train[column].isnull().values.sum()
    if missing > 0:
        no_cols += 1
print 'Total no. of cols: ', i
print 'No. of cols with missing values: ', no_cols, 'columns.'
```

```python
# One hot encoding for the categorical features.
train = pd.get_dummies(data=train, columns=cat_feat)
```
